```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load in data
BASE_URL = "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/"
files = {
    "demographics": "DEMO_J.XPT",
    "cholesterol": "TCHOL_J.XPT",
    "hdl": "HDL_J.XPT",
    "blood_pressure": "BPX_J.XPT",
    "blood_cholesterol": "BPQ_J.XPT",
    "diabetes_q": "DIQ_J.XPT",
    "glucose": "GLU_J.XPT",
    "smoking": "SMQ_J.XPT",
    "body_measures": "BMX_J.XPT"
}

dfs = {}
for name, filename in files.items():
    dfs[name] = pd.read_sas(BASE_URL + filename)
```

```python
dfs['blood_cholesterol'].head()
```

```python
# Merge dataframes on SEQN ID column
df_merged = dfs['demographics']
for name in dfs.keys():
    if name == 'demographics':
        continue
    df_merged = df_merged.merge(dfs[name], on='SEQN', how='inner')
```
